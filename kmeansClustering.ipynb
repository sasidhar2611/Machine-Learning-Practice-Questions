```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture

# Load your dataset (replace filename as needed)
df = pd.read_csv('sine_data.csv')

# K-Means Clustering
kmeans = KMeans(n_clusters=3)
clusters_kmeans = kmeans.fit_predict(df)
plt.scatter(df.iloc[:,0], df.iloc[:,1], c=clusters_kmeans)
plt.title('K-Means Clustering')
plt.show()

# Gaussian Mixture Model
gmm = GaussianMixture(n_components=3)
clusters_gmm = gmm.fit_predict(df)
plt.scatter(df.iloc[:,0], df.iloc[:,1], c=clusters_gmm)
plt.title('Gaussian Mixture Model Clustering')
plt.show()

# Hierarchical Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
clusters_agglo = agg_clustering.fit_predict(df)
plt.scatter(df.iloc[:,0], df.iloc[:,1], c=clusters_agglo)
plt.title('Hierarchical Clustering')
plt.show()
```
