```python
from hmmlearn import hmm
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load sequential data
data = pd.read_csv('sine_data.csv')

# Use 'y' column as sequential feature
sequence = data[['y']].to_numpy()

# Fit Gaussian HMM
model = hmm.GaussianHMM(n_components=2, n_iter=100, random_state=42)
model.fit(sequence)

# Predict hidden states
states = model.predict(sequence)
print('Predicted hidden states:', states)
```
