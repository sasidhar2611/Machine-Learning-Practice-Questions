```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('sine_data.csv')

# Create binary target from y
data['target'] = (data['y'] > 0).astype(int)
X = data[['x', 'y']]
y = data['target']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Decision Tree
cart = DecisionTreeClassifier(random_state=42)
cart.fit(X_train, y_train)

print('Test accuracy:', cart.score(X_test, y_test))

# Optimized tree plot
plt.figure(figsize=(6, 4))  # Smaller figure for small trees
plot_tree(cart, 
          filled=True, 
          feature_names=['x', 'y'], 
          class_names=['Negative', 'Positive'], 
          fontsize=12,               # Larger font for readability
          rounded=True,              # Rounded boxes look cleaner
          proportion=False,          # Display raw counts
          precision=2                # Control number formatting
         )
plt.tight_layout()
plt.show()
```
